```python
import pandas as pd
import numpy as np
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.corpus import stopwords
nltk.download('wordnet')
import re
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer() 
```

```python
df = pd.read_csv("questions.csv")
df1 = df["question1"]
```

```python
null_values = df1.isna().sum()/df1.count()
df1 = df1.astype("string")
```

```python
def preprocess(sentence):
    sentence=str(sentence)
    sentence = sentence.lower()
    sentence=sentence.replace('{html}',"")
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', sentence)
    rem_url=re.sub(r'http\S+', '',cleantext)
    rem_num = re.sub('[0-9]+', '', rem_url)
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(rem_num) 
    filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords.words('english')]
    stem_words=[stemmer.stem(w) for w in filtered_words]
    lemma_words=[lemmatizer.lemmatize(w) for w in stem_words]
    return " ".join(filtered_words)
df1= df1.iloc[0:10000].map(lambda s:preprocess(s))
df1 = df1.dropna()
```

```python
strings = df1.values
tokenized_sent = []
sentences = strings[0:10000]
for s in sentences:
    tokenized_sent.append(word_tokenize(s.lower()))
tokenized_sent
```

```python
# import
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
tagged_data = [TaggedDocument(d, [i]) for i, d in enumerate(tokenized_sent)]
tagged_data
```

```python
## Train doc2vec model
model = Doc2Vec(tagged_data, vector_size = 20, window = 2, min_count = 1, epochs = 100)
'''
vector_size = Dimensionality of the feature vectors.
window = The maximum distance between the current and predicted word within a sentence.
min_count = Ignores all words with total frequency lower than this.
alpha = The initial learning rate.
'''
## Print model vocabulary
```

```python
test_doc = word_tokenize("one earn money online without investment".lower())
test_doc_vector = model.infer_vector(test_doc)
results = model.docvecs.most_similar(positive = [test_doc_vector])
for i in results[0:4]:
    print( df1.iloc[i[0]] )
    print(i[1])
```
